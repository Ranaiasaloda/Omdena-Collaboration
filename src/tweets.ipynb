```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv(r'..\data\healthcare_tweets.csv')
sentiment_counts = df['sentiment'].value_counts()


sentiment_counts_df = sentiment_counts.reset_index()
sentiment_counts_df.columns = ['Sentiment', 'Count']


display(sentiment_counts_df)
```

```python

    plt.figure(figsize=(6,4))
    plt.bar(sentiment_counts_df['Sentiment'], sentiment_counts_df['Count'], color=['green','red','blue'])
    plt.title("Tweet Sentiments - Matplotlib")
    plt.xlabel("Sentiment")
    plt.ylabel("Number of Tweets")
    plt.savefig(r'..\plots\sentiment_bar.png')
    plt.close()     
```

```python
plt.figure(figsize=(6,6))

plt.pie(sentiment_counts,
        labels=sentiment_counts.index,
        autopct='%1.1f%%',
        colors=['green','red','blue'])

plt.title("Sentiment Percentage")

plt.savefig(r'..\plots\sentiment_pie.png')
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x='sentiment', data=df, palette='Set2')

plt.title("Sentiment Count")
plt.xlabel("Sentiment")
plt.ylabel("Tweets")

plt.savefig(r'..\plots\sentiment_count.png')
plt.show()
```

```python
from wordcloud import WordCloud

# Combine all tweets into a single string
all_tweets = " ".join(df['tweet'].astype(str))

# Create WordCloud object
wordcloud = WordCloud(width=800, height=400,
                      background_color='white',
                      colormap='Set2').generate(all_tweets)

# Plot WordCloud using Matplotlib
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Most Frequent Words in Tweets", fontsize=16)

# Save WordCloud to 'plots' folder
plt.savefig(r'../plots/wordcloud_tweets.png')
plt.close()
```
